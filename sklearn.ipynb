```python
# Your code goes here
print(__doc__)

# Based on Example from:
# Author: Noel Dawe <noel.dawe@gmail.com>
#
# License: BSD 3 clause

# importing necessary libraries
import numpy as np
import matplotlib.pyplot as plot
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor

# Create the dataset from random values
rng = np.random.RandomState(2)

# Set line spacings (start, stop, num=50, endpoint=True, retstep=False, dtype=None)
X = np.linspace(0, 10, 100)[:, np.newaxis]

# Sets y as the sin of x, ravel Returns a contiguous flattened array, set the distribution
Y = np.sin(X).ravel() + np.sin(10 * X).ravel() + rng.normal(0, 0.1, X.shape[0])

# Fit regression model
decision_tree_regression = DecisionTreeRegressor(max_depth=4)

ada_boost_regression = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4),
                          n_estimators=300, random_state=rng)

decision_tree_regression.fit(X, Y)
ada_boost_regression.fit(X, Y)

# Predict y value based on the x using the model
decision_tree_prediction_y = decision_tree_regression.predict(X)
ada_boost_prediction_y = ada_boost_regression.predict(X)

# Plot the results & add labels for the key/legend
plot.figure()
plot.scatter(X, Y, c="b", label="training samples")
plot.plot(X, decision_tree_prediction_y, c="g", label="n_estimators=1", linewidth=2)
plot.plot(X, ada_boost_prediction_y, c="r", label="n_estimators=300", linewidth=2)

# Add labels/title
plot.xlabel("random data")
plot.ylabel("target data")
plot.title("Example Boosted Decision Tree Regression")
plot.legend()
plot.show()
```
